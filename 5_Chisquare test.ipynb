#### Objective:
    To use the Chi-Square test for independence to determine if there's a significant association between the type of smart home device purchased (Smart Thermostats vs. Smart Lights) and the customer satisfaction level.


##### 1. State the Hypotheses:
###### Null Hypotheses(H0): there is no significant association between the type of smart home device purchase and the customer satisfaction level
###### Alternative Hypotheses(H1): there is significant association between the type of smart home device purchase and the customer satisfaction level

##### 2. Compute the Chi-Square Statistic:

```python
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
data = [
    [50,70], # very satisfied
    [80,100], # satisfied
    [60,90], # netural
    [30,50], # unsatisfied
    [20,50] # very unsatisified
]
data
```

```python
df = pd.DataFrame(data,index=['very satisfied','satisfied','netural','unsatisfied','very unsatisified'],columns=['smart thermostats','smart lights'])
df
```

```python
chi2_stat,p,dof,expected = chi2_contingency(df)
```

```python
print('chi-square statistic:',chi2_stat)
print('p-values:',p)
print('degrees of freedom:',dof)
print('expected frequencies:',expected)
```

##### 3. Determine the Critical Value:

```python
from scipy.stats import chi2

alpha = 0.05
critical_values = chi2.ppf(1 - alpha,dof)
critical_values
```

##### 4. Make a Decision:

```python
if chi2_stat > critical_values:
    print('conclusion: reject null hypothesis -> there is significant association exists')
else:
    print('conclusion: fail to reject null hypthesis -> there is no significant association')
```
