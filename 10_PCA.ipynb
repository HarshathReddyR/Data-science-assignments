### Dimensionality Reduction with PCA:

#### Standardize the features 

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

#### Implement PCA 

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
pca_data = pca.fit_transform(scaled_data)
```

#### Determine the optimal number of principal 

```python
explained_var = pca.explained_variance_ratio_
cumulative_var = explained_var.cumsum()
```

### Scree plot 

```python
plt.figure(figsize=(8,5))
plt.plot(range(1,len(explained_var)+1),cumulative_var,marker='*',linestyle='-')
plt.title('cumulative explained variance by PCA components')
plt.xlabel('number of principal components')
plt.ylabel('cumulative explained variance')
plt.show()
```

#### Choose no.of components

```python
n_comp = next(i for i,total in enumerate(cumulative_var) if total >= 0.95) + 1
print(f'optimal no.of components:{n_comp}')
```

#### Transform the original dataset into the principal components

```python
pca_final = PCA(n_components = n_comp)
redu_data = pca_final.fit_transform(scaled_data)
```

```python
# creating dataframe of principal components

pca_df = pd.DataFrame(redu_data,columns=[f'PC{i+1}' for i in range(n_comp)])
print(pca_df.head())
```

### Clustering with Original Data:

#### Apply clustering algorithm  KMeans

```python
from sklearn.cluster import KMeans
```

```python
# try different values of k

inertia = []
k_values = range(2,11)
for k in k_values:
    kmeans = KMeans(n_clusters=k,random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)
```

```python
# plotting elbow curve

plt.figure(figsize=(8,5))
plt.plot(k_values,inertia,marker='*',linestyle='-')
plt.title('elbow method for optimal k')
plt.xlabel('number of clusters (k)')
plt.ylabel('inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3,random_state=42)
cluster = kmeans.fit_predict(scaled_data)
```

```python
data['cluster'] = cluster
print(data.head())
```

### Visualize the clustering 

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=data.iloc[:,0],y=data.iloc[:,1],hue=data['cluster'],palette='Set2',s=80)
plt.title('KMeans clustering on original data')
plt.xlabel(data.columns[0])
plt.ylabel(data.columns[1])
plt.legend(title='cluster')
plt.show()
```

### Evaluate the clustering performance 

```python
from sklearn.metrics import silhouette_score,davies_bouldin_score
```

```python
sil_score = silhouette_score(scaled_data,cluster)
db_index = davies_bouldin_score(scaled_data,cluster)
```

```python
print(f'silhoette score:{sil_score:.3f}')
print(f'davies bouldin score:{db_index:.3f}')
```

### Clustering with PCA Data:

#### Apply clustering algorithm KMeans

```python
kmeans_pca = KMeans(n_clusters=3,random_state=42)
cluster_pca = kmeans_pca.fit_predict(pca_data)
```

### Visualize the clustering

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=pca_data[:,0],y=pca_data[:,1],hue=cluster_pca,palette='Set2',s=80)
plt.title('KMeans clustering on PCA transformed data')
plt.xlabel('principal component 1')
plt.ylabel('principal component 2')
plt.legend(title='cluster')
plt.show()
```

### Evaluate the clustering performance

```python
sil_score_pca = silhouette_score(pca_data,cluster_pca)
db_index_pca = davies_bouldin_score(pca_data,cluster_pca)
```

```python
print(f'silhoette score pca:{sil_score_pca:.3f}')
print(f'davies bouldin score pca:{db_index_pca:.3f}')
```

### Comparison and Analysis:

```python
### Both original and PCA datasets  gave identical clustering results
### The clustering patters remained the same, showing PCA preserved the data structure
### PCA reduced dimensionality without affecting clustering performance
### PCA improved visualization and computation speed while maintaining accuracy
### The trade off is reduced feature interpretability but easier analysis and clearer cluster seperation
### PCA is effective for simplifying data before clustering with minimal performance loss
```
