#### 1.Prime number

```python
n = int(input('Enter the number : '))
# Function to check if a number is prime
def prime_number(n):
    count = 0 # creating and assigning the 0 to count for the and track the count of divisor 
    if(n<=1): # Prime numbers are greater than 1, so 0, 1, and negatives are not prime
        return f'{n} is not a prime number'
    if(n==2): # 2 is only even prime number
        return f'{n} is a prime number'
    if(n%2 == 0): # even number are not prime number excepted 2
        return f'{n} is not a prime number'
    for i in range(3,n,2): # checking for the odd numbers
        if(n%i == 0): # if divisable increment the count
            count += 1
        if(count > 2): # if divisor is more than 2 not a prime
            return f'{n} is not a prime number'
    return f'{n} is a prime number'# if no divisor is find then it is prime number

prime_number(n)
    
```

#### 4.Word counter
write a program to count the number of words in a given text.
example:
input_text = "This is a sample text. This text will be used to demonstrate the word counter."
Expected output:
'This': 2 
'is': 1
'a': 1
'sample': 1
'text.': 1


```python
# creating a function for the word counter 
def word_counter(text):
    special_chars = [ " ", "\t", "\n", ".", ",", ";", ":", "?", "!",  "\"","(", ")", "[", "]", "{", "}", "-", "_", "/", "\\"]
    words={}
    word=""
    for ch in text:
        if ch not in special_chars:  # forming the word untill special_chars
            word+=ch
        else:                        # if any special_chars are appere
            if word != "":           
                if word in words:    # if word is already in the words increment the count
                    words[word] +=1
                else:                # word is not there in the words add to words with one count
                    words[word] =1
                word=""              # reseting the word for to find the new word

    if word != "":                   # if input is only one word or to get one word
        if word in words:
            words[word] += 1
        else:
            words[word] = 1

    for w, c in words.items():       # displaying the words items 
        print(f"{w.ljust(15)}: {c}") # ljust() is used for uniform display ljust() make the word length 15 if less it fill with spaces


sentence= input("Give any sentence : \n")
word_counter(sentence)
```

#### 5.Check for Palindrome
Write a Python function called is_palindrome that takes a string as input and returns True if the string is a palindrome, and False otherwise. A palindrome is a word, phrase, number, or other sequence of characters that reads the same forward and backward, ignoring spaces, punctuation, and capitalization.


```python
# creating a function for to chech the palindrome with time "Time Complexity O(n) "
def is_palindrome(a):
    s=a.lower() # converting the input into lower 
    if s == s[::-1]: # comparing the s with reverse string
        print(f" '{a}' is palindrome.")
    else:
        print(f" '{a}' is not a palindrome.")

word = input("Enter the word to check for palindrome\n")
is_palindrome(word)
```
